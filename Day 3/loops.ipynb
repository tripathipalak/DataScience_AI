```python
# pattern printing
for i in range(1,6):
    for j in range(1,6):
        print("*", end=" ")  
    print()
```

```python
n = 5
fact = 1
i = 1
while(i<=n):
    fact = fact*i
    i += 1
print("factorial is:",fact)    
```

```python
def add(a, b):
    """returns the sum of two numbers."""
    return a + b

result = add(10, 20)
print("Addition:",result)
```

```python
def sub(a, b):
    return a - b

result = sub(20, 10)
print("Substraction:",result)

def sub(a, b):
    return a * b

result = sub(20, 10)
print("Multiplication:",result)

def sub(a, b):
    return a / b

result = sub(20, 10)
print("Division:",result)
```
